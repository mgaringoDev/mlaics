# Renaming

## rename()

We can use the rename method to rename the coloumn.  It takes in a **coloumn dictionary** indexed with the current coloumn name and valued to the renamed column name.

```python
coloumns={'currentColName':'changedColName'}
```

````python
reviews.rename(columns={'points':'score'})
````

The above changed the name **points** to **score**.  You can also do this for the index.

## Index renaming

````python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
````

# Combining

When performing operations on a dataset, we will sometimes need to combine different DF and/or Series in non-trival ways.  Pandas has three cre methods **concat()** **join()** **merge()**.  Most of what **merge()** can do can also be done more simply with **join()**, we we will omit it and focus on the **concat()** and **join()**

## concat()

Given a list of elements, this fucntion will smush those elements togther along an axis.  This is useful when we have data in different DF or Series objects but having the same fields.  

Lets use the [Youtube Videos DB](https://www.kaggle.com/datasnaek/youtube-new#GBvideos.csv) as an example which splits the data up based on country of origin.  Now let us study multiple countries simultaneously we can use **concat()**.

````python
canadian_youtube = pd.read_csv("../data/CAvideos.csv")
british_youtube = pd.read_csv("../data/GBvideos.csv")
````

````python
canadian_youtube.head()
````

````python
british_youtube.head()
````

Notice that they have the same coloumns but different data.  We can concatnate these two using the function **concat()**, which is a vertical concatination.

````python
pd.concat([canadian_youtube, british_youtube])
````

## join()

The middle most combiner in terms of complexity is **join()**.  This lets us combine different DF objects which have an index in common.  For example, to pull down videos that happend to be trending on the same day in BOTH Can and UK we can do the following.


The **lsuffix** and **rsuffix** parameters are **necessary here because the data has the same column names** in both British and Canadian datasets. If this wasn't true (because, say, we'd renamed them beforehand) we wouldn't need them.

````python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
````
