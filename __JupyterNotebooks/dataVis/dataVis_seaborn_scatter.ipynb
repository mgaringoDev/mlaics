# Colour Coded

```python
plt.figure(figsize=(14,4))

sns.scatterplot(x=dfInsurance['bmi'],
               y=dfInsurance['charges'],
               hue=dfInsurance['smoker'])

plt.show()
```

So you colour the scatter plot based on data in smoker

## colour coded with regression **lmplot()**

```python
plt.figure(figsize=(14,4))

sns.lmplot(x='bmi',
          y='charges',
          hue='smoker',
          data=dfInsurance)

plt.show()
```

Notice that data should be a DF and the inputs x,y, and hue are the coloumn values to be highlighted.

## swarmplot()

This is often refered to as **categorical scatter plot**.

```python
plt.figure(figsize=(10,4))

sns.swarmplot(x=dfInsurance['smoker'],
             y=dfInsurance['charges'])

plt.show()
```
