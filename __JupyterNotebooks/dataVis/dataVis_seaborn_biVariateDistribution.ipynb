# Imports

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# I/O

```python
irisFileName = "../data/iris.csv"
```

```python
dfIris = pd.read_csv(irisFileName,
                         index_col='Id')
```

```python
dfIris.head()
```

# Data

We will be plotting a bunch of stuff so based on the above dataset lets only look at two coloumns and make those the variables of interest for analysis.

```python
x = dfIris['Sepal Length (cm)']
y = dfIris['Petal Length (cm)']
```

```python
dictData = {'x':x,
           'y':y}
dfData = pd.DataFrame(dictData)
```

```python
dfData.head()
```

# Plotting **jointplot()**

The main function we will be using is called the **[jointplot()](https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot)**.  Which basically creates the two variable distributions so we can see both in space and time.

> **kind{ “scatter” | “reg” | “resid” | “kde” | “hex” }**, optional
Kind of plot to draw.

## Scatter

```python
plt.figure(figsize=(14,4))


sns.jointplot(x="x",
             y="y",
             data=dfData)

plt.show()
```

## Hexbin

```python
plt.figure(figsize=(14,4))


sns.jointplot(x="x",
             y="y",
             data=dfData,
             kind='hex')

plt.show()
```

## Hex with scatter

```python
plt.figure(figsize=(14,4))

# first draw the kde
g = sns.jointplot(x="x",
             y="y",
             data=dfData,
             kind='hex')

# adding scatter plot to the grid
g.plot_joint(plt.scatter,c='r',marker='x')

plt.show()
```

# KDE

```python
plt.figure(figsize=(14,4))


sns.jointplot(x="x",
             y="y",
             data=dfData,
             kind='kde')

plt.show()
```

## KDE with Scatter

```python
plt.figure(figsize=(14,4))

# first draw the kde
g = sns.jointplot(x="x",
             y="y",
             data=dfData,
             kind='kde')

# adding scatter plot to the grid
g.plot_joint(plt.scatter,c='r',marker='x')

plt.show()
```
