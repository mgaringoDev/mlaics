# Imports

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# I/O

```python
spotifyFilepath = "../data/spotify.csv"
```

```python
dfSpotify = pd.read_csv(spotifyFilepath,
                        index_col="Date",
                       parse_dates=True)
```

```python
dfSpotify.head()
```

```python
dfSpotify.tail()
```

There are some NaN's in the dataset.  So this data set is very dirty.  Lets clean it up later.

```python
dfSpotify = dfSpotify.fillna(0)
```

# Plotting

```python
width = 14 # inches
height = 6 # inches
plt.figure(figsize=(width,height))

plt.title('Top Songs')

sns.lineplot(data=dfSpotify)

plt.show()
```

## Plotting Subset data

Lets say you only want to plot **Shape of you** we can do the follow.

```python
dfSpotify.columns
```

```python
width = 14 # inches
height = 6 # inches
plt.figure(figsize=(width,height))

plt.title('Top Songs')

sns.lineplot(data=dfSpotify[['Shape of You','HUMBLE.']],
             label=['Eddy She','Kenddy'])
#sns.lineplot(data=dfSpotify['HUMBLE'],label=)

plt.show()
```
