# Grid Usage

I think I like using **plt.subplot2grid** more than **plt.subplot()** because of its ease of customizability.  But both is fine.

```python
plt.figure(figsize=(14,4))

# __GRID___ tuple with (R,C) number of rows and coloumn
# _POSITION_ tuple with (R,C) for starting position to draw
# __ROWSPAN__ number of rows to take
# __ROWSPAN__ number of cols to take

ax1 = plt.subplot2grid((_GRID_),(_POSITION_),rowspan=__ROWSPAN__, colspan=__COLSPAN)

plt.show()
```

````python
plt.figure(figsize=(14,4))

# __GRID___ tuple with (R,C) number of rows and coloumn
# _POSITION_ tuple with (R,C) for starting position to draw
# __ROWSPAN__ number of rows to take
# __ROWSPAN__ number of cols to take

ax1 = plt.subplot2grid((2,3),(0,0),rowspan=1, colspan=1)
# DRAW FIRST GRAPH
sns.scatterplot(x='Sepal Length (cm)',
               y='Sepal Width (cm)',
                hue='Species',
               data=dfData)
ax1.set_title('Sepal')


ax2 = plt.subplot2grid((2,3),(0,2),rowspan=1, colspan=1)
# DRAW SECOND GRAPH
sns.scatterplot(x='Petal Length (cm)',
               y='Petal Width (cm)',
                hue='Species',
               data=dfData)
ax2.set_title('Petal')


ax3 = plt.subplot2grid((2,3),(1,0),rowspan=1, colspan=3)
# DRAW THIRD GRAPH
sns.scatterplot(x='Petal Length (cm)',
               y='Sepal Width (cm)',
                hue='Species',
               data=dfData)
ax3.set_title('Petal vs Sepal')



plt.show()
````

Since this looks very cluttered we can even add padding using **subplots_adjust(wspace=0.5, hspace=0.5)** [subplot_adjust()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

````python
plt.figure(figsize=(20,10))
plt.subplots_adjust(wspace=0.5, hspace=0.5)


# __GRID___ tuple with (R,C) number of rows and coloumn
# _POSITION_ tuple with (R,C) for starting position to draw
# __ROWSPAN__ number of rows to take
# __ROWSPAN__ number of cols to take

ax1 = plt.subplot2grid((2,3),(0,0),rowspan=1, colspan=1)
# DRAW FIRST GRAPH
sns.scatterplot(x='Sepal Length (cm)',
               y='Sepal Width (cm)',
                hue='Species',
               data=dfData)
ax1.set_title('Sepal')


ax2 = plt.subplot2grid((2,3),(0,2),rowspan=1, colspan=1)
# DRAW SECOND GRAPH
sns.scatterplot(x='Petal Length (cm)',
               y='Petal Width (cm)',
                hue='Species',
               data=dfData)
ax2.set_title('Petal')


ax3 = plt.subplot2grid((2,3),(1,0),rowspan=1, colspan=3)
# DRAW THIRD GRAPH
sns.scatterplot(x='Petal Length (cm)',
               y='Sepal Width (cm)',
                hue='Species',
               data=dfData)
ax3.set_title('Petal vs Sepal')



plt.show()
````

# Arrows and Annotations

Drawing arrows in Matplotlib is often much harder than you'd bargain for. While there is a plt.arrow() function available, I wouldn't suggest using it: the arrows it creates are SVG objects that will be subject to the varying aspect ratio of your plots, and the result is rarely what the user intended. Instead, I'd suggest using the **plt.annotate()** function. 

````python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"));
````

**arrowprops** takes in a dictionary in which you can manipulate the arrow object in what ever way you want.  For more examples refer to the [doc](https://matplotlib.org/examples/pylab_examples/annotation_demo2.html)
